```python
import pandas as pd
```

```python
data = [
    {"Title": "Python Developer", "Company": "TCS", "Location": "Hyderabad", "Experience": "0-2", "Salary_LPA": 5, "Remote": "No"},
    {"Title": "Senior Python Engineer", "Company": "Infosys", "Location": "Hyderabad", "Experience": "3-5", "Salary_LPA": 10, "Remote": "No"},
    {"Title": "Java Developer", "Company": "Wipro", "Location": "Bangalore", "Experience": "0-2", "Salary_LPA": 6, "Remote": "Yes"},
    {"Title": "Data Analyst", "Company": "HCL", "Location": "Chennai", "Experience": "1-3", "Salary_LPA": 7, "Remote": "No"},
    {"Title": "Backend Developer", "Company": "Startup XYZ", "Location": "Hyderabad", "Experience": "2-4", "Salary_LPA": 8, "Remote": "Yes"},
    {"Title": "Machine Learning Engineer", "Company": "Tech Mahindra", "Location": "Bangalore", "Experience": "0-2", "Salary_LPA": 9, "Remote": "No"},
    {"Title": "Frontend Developer", "Company": "Cognizant", "Location": "Mumbai", "Experience": "0-2", "Salary_LPA": 6, "Remote": "Yes"},
]

df = pd.DataFrame(data)
df
```

```python
def search_jobs(role, location):
    results = df[
        (df["Title"].str.contains(role, case=False)) &
        (df["Location"].str.contains(location, case=False))
    ]
    return results
```

```python
search_jobs("python", "Hyderabad")
```

```python
def filter_by_salary(df, max_salary):
    return df[df["Salary_LPA"] <= max_salary]
```

```python
results = search_jobs("python", "Hyderabad")
filter_by_salary(results, 8)
```

```python
def filter_by_experience(df, exp_range):
    return df[df["Experience"] == exp_range]
```

```python
role = input("Enter Job Role: ")
location = input("Enter Location: ")
max_salary = int(input("Enter Max Salary (LPA): "))

results = search_jobs(role, location)
results = filter_by_salary(results, max_salary)

results
```
